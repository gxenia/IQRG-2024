### ***The scenario:***

Two parties, commonly named Alice and Bob, want to teleport a quantum state ∣ψ⟩ from Alice to Bob ∣ψ⟩

Alice and Bob share an entangled pair of particles, with one particle held by Alice (particle A) and the other by Bob (particle B). the particles are in the bell states(above).

Alice performs a joint quantum measurement on her particle A (part of the entangled pair) and the particle whose state ∣ψ⟩ she wants to teleport (let’s call it particle C).

This measurement entangles particles A and C, collapsing their joint state into one of the four possible Bell states. This measurement process can be represented mathematically and involves projecting the state onto one of the Bell basis states.

After performing the measurement, Alice obtains one of four possible outcomes. She sends the result of her measurement to Bob using classical communication (two classical bits of information).

The four possible outcomes correspond to the four Bell states, and they inform Bob about the specific transformation needed to retrieve the original state.

Based on the two bits of classical information received from Alice, Bob applies a specific unitary transformation (Pauli operators: I, X, Y, or Z) to his particle B.

These operations correct the state of particle B so that it matches the original state ∣ψ⟩ of particle C that Alice wanted to teleport.

*4 possible bell states of particles:*

|Φ+⟩ = (|00⟩ + |11⟩)/√2

 |Φ-⟩ = (|00⟩ - |11⟩)/√2

|Ψ+⟩ = (|01⟩ + |10⟩)/√2

 |Ψ-⟩ = (|01⟩ - |10⟩)/√2


The probability for each of the four possible measurement outcomes (00,01,10,11) is ¼

### **Code:**


```python
from qiskit import *

#creating circuit and apply X gate on 1st qubit
circuit = QuantumCircuit(3, 3)
circuit.x(0)
circuit.barrier()

#apply Hadamard gate and Controlled-X gate on corresponding qubits
circuit.h(1)
circuit.cx(1, 2)
circuit.cx(0, 1)
circuit.h(0)
circuit.barrier()

# measure it
circuit.measure([0, 1], [0, 1])
circuit.barrier()

#finally apply Controlled-X  gate and Controlled-Z  gate on corresponding qubits
circuit.cx(1, 2)
circuit.cz(0, 2)

#draw the whole circuit
circuit.draw(output='mpl', style = 'iqp')
```

Aer and Qiskit Plot histogram components:

```python
#measure out from the first 2 Qubits and 2 Classical Bits
circuit.measure(2, 2)

simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit, backend = simulator, shots =1024).result()
counts = result.get_counts()
from qiskit.tools.visualization import plot_histogram
plot_histogram(counts)
```
